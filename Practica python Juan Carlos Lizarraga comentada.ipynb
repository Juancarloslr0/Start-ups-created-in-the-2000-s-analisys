# Ejercicio 9. 
##  Crear un nuevo subconjunto filtrando del dataset todas las filas por debajo de 10 millones y visualizar el histograma.
##  RAISED_AMOUNT es la columna que queremos filtrar por debajo de los 10 millones.

```python
# Asigno la variable "subconjunto" y con el siguiente codigo indico que necesito los valores por debajo de 10 millones de la columna raised_amount.
subconjunto = dataframe_startups1[dataframe_startups1['raised_amount'] < 10000000]
subconjunto
```

```python
plt.hist(subconjunto['raised_amount'], bins=20, color='blue', alpha=1)
plt.title('Histograma  de  start ups ,  en las que se generó menos de $10,000,000 USD y su frecuencia')
plt.xlabel('Cantidad generada')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.show()
```

```python
## 10. Con el conjunto del apartado anterior, investigar sobre la función “contains” y filtrar el dataset por 
## las fechas de creación de las startup en 2000
```

```python
# Asigno la función subset_2000 para trabajar el código, y la funcion .loc, indico que quiero aplicar filtro en la columna "funded_at",
# dentro de eso filtro, uso la funcion contains, que usa la máscara booleana, indicandole na=False, y anteponiendo "2000" del año que queremos analizar,
# obtendremos los datos de las start ups fundadas en el 2000. 
subset_2000 = subconjunto.loc[subconjunto['funded_at'].str.contains('2000', na=False)]
subset_2000
```

# Ejercicio 11. 
## Sobre el conjunto filtrado por fecha, visualizar la cuantía obtenida explicada por el tipo de moneda.

```python
# Asigno variable para trabajar, dentro del subconjunto anterior, le indico las columnas que hay que analizar, las voy a cruzar con la funcion sum,
# y así poder visualizarlas, de ahi podemos sacar concluciones, explicar lo ocurrido en esos años con las start ups, etc, gracias al análisis de datos.
subset_cantidad_vs_divisa_2000 = subset_2000.groupby('raised_currency_code')['raised_amount'].sum()
subset_cantidad_vs_divisa_2000
```

```python
# Las visualizamos de la siguiente manera, personalizo mi gráfica, y así podemos exponer los hechos a cualquier público
subset_cantidad_vs_divisa_2000.plot(kind='bar')
plt.title('Cantidad generada por cada una de las Divisas en los años 2000')
plt.xlabel('Divisa')
plt.ylabel('Numero de start ups')
plt.grid(True)

plt.show()
```
